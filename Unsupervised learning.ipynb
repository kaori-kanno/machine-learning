Unsupervised learning:

    Aquí el algoritmo está buscando patrones entre los datos, por tanto nosotros tendremos que tratar de representar la data de forma simple y reducida, una representación comprimida de K dimensiones de nuestro total de datos N, K <<N.
    
    Algunas formas de hacerlo son:

Principal Component Analysis:

- Cuando los datos tienen una gran correlación entre ellos, podemos decir que quizá no necesitamos todos estos datos, sino solo algunos que representen los otros, es decir de nuestros datos N-dim hacemos una matriz de eigenvectors K-dimensionales. Se hace esta transformación en el espacio de los features.

K-means clustering:

- En este caso tenemos nuestros puntos, y elegimos ciertos "centroides", que representarán los datos más cercanos a estos.

Autoencoders:

Un autoencoder es un neural network entrenado para generar un código que represente nuestra data y que se pueda decodificar en el input inicial, es decir el output va a ser este mismo input. El código entonces nos dará la misma información, pero comprimida, de tal forma que sea más simple procesarla. Sirve sobretodo para trabajar con imágenes.
