Introducción:
    
    Si queremos estimar o predecir algo, no conocemos una regla exacta que nos entregue el resultado, pero sabemos que debería existir alguna regla que nos permita predecir el resultado. Si tenemos un montón de datos, sabemos que al usarlos de input deberá existir un output de estos, que será buscar una relación entre todos estos.

Programar v/s aprender:
    
    Usualmente al programar, le entregamos a un programa una serie de reglas, de tal forma que la computadora genere un resultado a partir de estos. Sin embargo, si queremos que el programa aprenda, queremos que este sea capaz de generar sus propias reglas, y que las vaya perfeccionando, que al agregar nueva información vaya optimizando estas reglas y reajustándolas.
    
    Estaoms aquí buscando tendencias entre los datos (colores, precios, información que relacione y diferencie los datos). Esto será lo que usaremos de input, los atributos de la información que estamos usando. Toda la información que pueda ser representada numéricamente, va a funcionar bien.
    
precisión o robusticidad:

    Un programa preciso será aquel que sea muy bueno en una sola específica área, muy específica, sin embargo, al meter este programa en otra área, será incapaz de completar la tarea que se le pida, a menos que se le reentrene para ello. Un programa robusto en cambio, será menos preciso, pero más general, y podrá aplicarse a más áreas (Ej: un programa que diferencie entre alimentos de distintos tipos, y otro que diferencia entre distintos alimentos y distintos camiones).
    
    Ahora, ningún programa será tan preciso, siempre habrá algo de aleatoriedad que no estamos representando correctamente, lo que llevará a errores en la estimación. El output siempre será más un aproximado del resultado correcto.

Existen otros modos de generar el error que pueden ser mejores o peores según la distribución de datos que tengamos (gaussiana, laplaciana, etc):

- Mean squared error, ê = (y - ỹ)^{2}: y has a Gaussian distribution around the expected (predicted) value ỹ
- Mean absolute error, ê = |y - ỹ|: y is Laplacian (two-sided exponential) around ỹ
- Cross entropy, ê = y Log(ỹ) + (1-y) Log(1-ỹ): y (y= 0 or 1) is a Bernoulli variable with < y >= ỹ

Para modelas estas funciones entonces debemos encontrar los parámetros que minimicen el error, lo cual no es fácil.

Tras haber entrenado nuestro programa y haber encontrado en error de cada uno de nuestros datos, el programa conoce ciertas reglas, por tanto ya no necesita la información con la que se le entrenó. Tener en cuenta además, que esto no significa que obtendremos necesariamente la información que nos interesa, bien podría ser que las relaciones encontradas den respuesta a algo completamente distinto a lo esperado.

Diferencia entre IA y machine learning:



Definición formal de machine learning:

    “A computer program is said to learn from experience E with respect to some class of tasks T and performance measure P, if its performance at tasks in T, as measured by P, improves with experience E”


Es diferente a la estadística en que esta última tiene el foco en qué tan probable es que ocurra algo, mientras que ML busca más bien predecir, diferenciar entre objetos o informaciones que usemos de inputs.

Se promedian los puntos (data) que estén cercanos unos del otro, y a los nuevos puntos a testear se les otorgará un valor cercano al círculo en el cual cae dentro de los datos de entrenamiento. Entonces, haremos un promedio (o se elegirá según la mayoría de datos que den un mismo resultado) entre los puntos de entrenamiento y estos nos entregarán la imagen y, es decir, la función es continua, y podremos separar los datos entre los que quedan dentro o fuera de ciertos círculos (o sobre o bajo cierta curva no necesariamente lineal) dentro de los cuales el algoritmo predijo que se tendrían ciertos valores y como resultado. Si la función es caótica, esto no tiene sentido. Aquí todos los datos son parámetros de la función.
