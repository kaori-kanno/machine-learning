Deep Learning:

    Es parte de las llamadas "artificial neural networks", algoritmos de machine learning, el cual es capaz de capturar la estructura de nuestra data, es decir es un procesamiento muy profundo. Trata de emular un cerebro, en el sentido en que al recibir información hace "sinapsis", tiene muchas conexiones distintas, y tiene un comportamiento complejo.


    Tendremos entonces varios inputs $I_{i}$, cada uno con un peso $w_{i}$ distinto, e ingresando todos estos y aplicando distintas funciones logramos determinar un output y con toda esta data. Mientras más conexiones se puedan hacer, más inteligente se considera al algoritmo, mientras este no quede saturado.




    La teoría detrá de esto la llamamos "backpropagation", creamos funciones que generen nuestros resultados. En esta teoría el error es la derivada parcial. Como sabemos esta nos puede ayudar a optimizar por definición de derivada, y añadimos este error a nuestra data.

Representation learning:

    Es una transofrmación de nuestras coordenadas de cada punto, de tal forma de poder separar la data de forma que quede más simple trazar una línea que separe los datos.
    



De esta forma con transformaciones muy simples podemos separar los datos.

Los algoritmos de Deel network funcionan mejor cuando tenemos que hacer tareas que son simples de hacer para una persona, pero que son difíciles de describir para estas; por ejemplo: reconocer un perro en una foto, v/s describir el proceso detrás que nos permite clasificar la foto como una de un perro. En este caso, el programa mide colores, tamaños, formas, etc. Pero nosotros lo hacemos de forma muy intuitiva y automática, tal que no lo notamos.

Computacionalmente, cada uno de estos pasos es igual para el algoritmo, y pueden pesar lo mismo.




El cómo estas funciones para definir edges, contours, etc. se generan, es automático, y difícilmente imitable por un humano.b

Si tenemos mucha data, con Deep Learning el algoritmo podrá extraer la información más relevante sin necesidad de comprimirla, pero tendremos que hacer otras cosas:

- Massaging the data:
      No necesitamos hacer mucho preprocessing, debido a que usamos deep learning y no hay que eliminar tanta data como en machine learning, pero nos sirve hacer ciertas transformaciones, como normalizar la data para ver cómo la data afecta al modelo, o pesar de ciertas formas los features. 
      También podemos hacer "data augmentation", es decir colocar la información que tenemos en los puntos para complementarlos y tener más información.
      
- Choosing your model: 
      Según la relación entre nuestros datos, buscamos un modelo que se adapte a estos. Los que hemos visto han sido LR y k-nearest. Para esto debemos colocar cierta data del task, de lo que queremos obtener, en el algoritmo. Aquí tenemos las activation functiones, CNN, transformers, etc.
     
- Choosing your objective: 
      y(x), la función, el resultado medible para entrenar nuestros parámetros. Es lo mismo que en machine learning.

- Training your model:
      Se necesita optimizar la data, lo cual es muy complicado. Para esto usamos la derivada. Si tenemos demasiada data, no podemos usarla toda para entrenamiento, pero se pueden usar subsets de data, de tal forma de gastar menos potencial computacional. Se pueden usar optimizadores como ADAM.

- Testing your model:
      Es lo mismo que en machine learning.
